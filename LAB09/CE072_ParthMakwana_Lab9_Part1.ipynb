**Hierarchical Clustering**

**Diabetese Dataset**

```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
diabetes = datasets.load_diabetes()
print("Features: ", diabetes.feature_names)
data_train, data_test, target_train, target_test = train_test_split(diabetes.data,diabetes.target, test_size=0.2, random_state=72)
```

```python
plt.scatter(data_train[:, 0], data_train[:, -1],)
plt.show()
```

```python
dendrogram(linkage(data_test, 'single'),orientation='top',distance_sort='descending',show_leaf_counts=True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc_model = AgglomerativeClustering(n_clusters = 2,affinity='euclidean',linkage='ward')
hc_model.fit_predict(data_test)
plt.scatter(x=data_test[:, 0], y=data_test[:, 1], c=hc_model.labels_, cmap="rainbow")
plt.show()
```

**Iris Dataset**

```python
iris = datasets.load_iris()
print("Features: ", iris.feature_names)
print("Labels:", iris.target_names)
data_train, data_test, target_train, target_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=72)
```

```python
plt.scatter(data_train[:, 0], data_train[:, -1])
plt.show()
```

```python
dendrogram(linkage(data_test, "single"), orientation="top", distance_sort="descending", show_leaf_counts=True)
plt.show()
```

```python
hc_model = AgglomerativeClustering(n_clusters = 2,affinity='euclidean',linkage='ward')
hc_model.fit_predict(data_test)
plt.scatter(x=data_test[:, 0], y=data_test[:, 1], c=hc_model.labels_, cmap="rainbow", s=50)
plt.show()
```
