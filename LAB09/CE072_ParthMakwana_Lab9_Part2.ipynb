**K-means Clustering**

**Diabetes Dataset**

```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
diabetes = datasets.load_diabetes()
print("Features: ", diabetes.feature_names)
data_train, data_test, target_train, target_test = train_test_split(diabetes.data,diabetes.target, test_size=0.2, random_state=145)
```

```python
plt.scatter(data_train[:, 0], data_train[:, -1])
plt.xlabel("X Coordinates")
plt.ylabel("Y Coordinates")
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=72).fit(data_train)
```

```python
print(kmeans.cluster_centers_)
print("\n",kmeans.labels_)
```

```python
plt.scatter(data_train[:, 0], data_train[:, -1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x')
plt.title("Data points and cluster centroids")
plt.show()
```

```python
print(silhouette_score(data_train, kmeans.labels_))
```

```python
from yellowbrick.cluster import KElbowVisualizer
model = KMeans(random_state=72)
visualizer = KElbowVisualizer(model, k=(2,6), metric='silhouette',timings=False)
visualizer.fit(data_train)
visualizer.poof()
```

**Iris Dataset**

```python
iris = datasets.load_iris()
print("Features: ", iris.feature_names)
print("Labels: ", iris.target_names)
```

```python
data_train, data_test, target_train, target_test = train_test_split(iris.data,iris.target, test_size=0.2, random_state=145)
plt.scatter(data_train[:, 0], data_train[:, -1])
plt.xlabel("x coordinates")
plt.ylabel("y coordinates")
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=72).fit(data_train)
```

```python
print(kmeans.cluster_centers_)
print("\n",kmeans.labels_)
```

```python
plt.scatter(data_train[:, 0], data_train[:, -1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c="red", marker="x")
plt.title("Data points and cluster centroids")
plt.show()
```

```python
print(silhouette_score(data_train, kmeans.labels_))
```

```python
model = KMeans(random_state=72)
visualizer = KElbowVisualizer(model, k=(2, 6), metric="silhouette", timings=False)
visualizer.fit(data_train)
visualizer.show()
```
