```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
y_pred = []
len_x = len(X_train[0])
w = []
b = 0.2
print(len_x)
```

```python
entries = len(X_train[:,0])
entries
```

```python
for weights in range(len_x):
  w.append(0)
w
```

```python
def sigmoid(z):
  return 1/(1 + np.exp(-z))
```

```python
def predict(input):
  z = np.dot(input, w) + b
  h= sigmoid(z)
  for i in range(len(h)):
    if(h[i]>=0.5):
      h[i]=1
    else:
      h[i]=0
  return h
```

```python
def loss_func(y, a):
  total_bce_loss = np.sum(-a * np.log(y) - (1 - a) * np.log(1 - y))
  m = y.shape[0]
  j = total_bce_loss /m
  return j
```

```python
dw = []
db = 0
J = 0
alpha = 0.1
for x in range(len_x):
  dw.append(0)
```

```python
for i in range(3000):
    z = np.dot(X_train, w) + b
    y_pred = sigmoid(z)
    l = loss_func(y_pred, y_train)
    dw = np.dot((y_pred-y_train).T, X_train)/X_train.shape[0]
    db = np.mean(y_pred-y_train)
    w = w - alpha * dw
    b = b - alpha* db
    print("Round:",i,"Weight:",w,"Bias:",b)
```

```python
y_pred=predict(X_train)
print(y_pred)
```

```python
print("actual Value\tpredicted values")
for i in range(len(y_test)):
    print(y_test[i],"\t\t",int(y_pred[i]))
```

```python
acc = np.sum([y_test[i] == int(y_pred[i]) for i in range(len(y_test))])/len(y_test)
print("Accuracy:",acc)
```
